```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('String Title Here')
plt.show()
```

# Basic Commands

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=12,alpha=0.5)
```

```python
fig =plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3.22,linestyle ='-',marker = 'o',markersize=11,markerfacecolor='yellow',
       markeredgecolor='green')
# can use ls instead of linestyle
```
